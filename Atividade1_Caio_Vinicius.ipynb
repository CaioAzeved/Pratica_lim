```python
import sympy
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
x, y, z = symbols('x y z')
init_printing(use_unicode=True)
```

```python
x1 = np.arange(-50, -2, 0.1)
x2 = np.arange(-1, 2, 0.1)
x3 = np.arange(2, 52, 0.1)
```

```python
y1 = x1 + 1
y1
```

```python
y2 = -(x2**2) + 2
y2
```

```python
y3 = x3
y3
```

```python
def plot_xy(x, y):
    fig, ax = plt.subplots(1,figsize=(10,5))
    ax.plot(x,y)
    ax.set_title('f(x)=???')
plot_xy(x1, y1)
```

```python
plot_xy(x2, y2)
```

```python
plot_xy(x3, y3)
```
